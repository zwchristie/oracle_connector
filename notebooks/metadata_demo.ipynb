```python
from app.config import Settings
from app.services.metadata import OracleMetadataService
from app.db import OracleClient

settings = Settings()
try:
    client = OracleClient(settings)
    service = OracleMetadataService(client, settings)
except RuntimeError as exc:
    service = None
    print(f"Oracle client is not available: {exc}")
```

```python
if service:
    filtered_tables = service.get_filtered_tables("HR")
    filtered_tables
else:
    print("Service not initialized. Configure Oracle/Kerberos and re-run the previous cell.")
```

```python
if service:
    manifest = service.generate_mdl(
        catalog="Analytics",
        schema="HR",
        tables=None,
        apply_usage_filter=True,
    )
    manifest
else:
    print("Service not initialized. Configure Oracle/Kerberos and re-run the initialization cell.")
```
